```python
## Örnek ses tasarımı ##

''' Aşağıdaki kod sonucu oluşan ses:
DO4-RE4-Mİ4-FA4-SOL4-LA4-Sİ4-Sİ4-LA4-SOL4-FA4-Mİ4-RE4-DO4 = C4-D4-E4-F4-G4-A4-B4-B4-A4-G4-F4-E4-D4-C4

'''

import sounddevice as sd
import numpy as np 
from math import pi
import matplotlib.pyplot as plt
from scipy.fftpack import fft , ifft
import wavio


fs = 44100

n = np.arange(0,0.4,1/fs) 


f = 261.626 #DO4 = C4
f2 = 293.665 #RE4 = D4
f3 = 329.628 #Mİ4 = E4
f4 = 349.228 #FA4 = F4
f5 = 391.995 #SOL4 = G4
f6 = 440.000 #LA4 = A4
f7 = 493.883 #Sİ4 = B4

tam_do = np.sin(2*pi*f*n) 
tam_re = np.sin(2*pi*f2*n) 
tam_mi = np.sin(2*pi*f3*n)
tam_fa = np.sin(2*pi*f4*n)
tam_sol = np.sin(2*pi*f5*n)
tam_la = np.sin(2*pi*f6*n)
tam_si = np.sin(2*pi*f7*n)


delay = np.zeros(4410)


tam_do_delay = np.append(tam_do,delay)
tam_re_delay = np.append(tam_re,delay)
tam_mi_delay = np.append(tam_mi,delay)
tam_fa_delay = np.append(tam_fa,delay)
tam_sol_delay = np.append(tam_sol,delay)
tam_la_delay = np.append(tam_la,delay)
tam_si_delay = np.append(tam_si,delay)

y = np.append(tam_do_delay,tam_re_delay)
y2 = np.append(y,tam_mi_delay)
y3 = np.append(y2,tam_fa_delay)
y4 = np.append(y3,tam_sol_delay)
y5 = np.append(y4,tam_la_delay)
y6 = np.append(y5,tam_si_delay)

z = np.append(tam_si_delay,tam_la_delay)
z2 = np.append(z,tam_sol_delay)
z3 = np.append(z2,tam_fa_delay)
z4 = np.append(z3,tam_mi_delay)
z5 = np.append(z4,tam_re_delay)
z6 = np.append(z5,tam_do_delay)


notalar = np.append(y6,z6)


sd.play(notalar,fs)




```

```python
## oluşturduğumuz ses dosyasını kodun çalıştığı klasörün içine kaydolması için bu bloğun çalıştırılması gerekmektedir ##

'''dilerseniz daha sonra ses dosyasının kaydolduğu konumu nota tahmini içindeki wavfile.read() fonksiyonu içine yazarak 
nota tahmini kısmında doğruluğunu gözlemleyebilirsiniz '''

wavio.write("notalar.wav",notalar, fs, sampwidth=2)
```
